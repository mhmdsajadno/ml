```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# data
x = np.array([1,2,3,4]).reshape(-1,1)
y = np.array([2,3,5,7]).reshape(-1,1)
x.shape
```

```python
# plot
plt.plot(x,y,".")
plt.show()
```

```python
# using model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x, y)
```

```python
# y = mx + c
m = model.coef_
c = model.intercept_
print(m,c)
```

```python
# plot model
plt.plot(x, y, ".")
plt.plot(x,model.predict(x))
plt.show()
```
